```python
revelent_cols = ["Age","Annual Income (k$)","Spending Score (1-100)"]
customers_df = df[revelent_cols]
```

```python
customers_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(customers_df)
```

```python
scaled_data = scaler.transform(customers_df)
```

```python
scaled_data
```

```python
def find_best_clusters(df, maximum_K):
    clusters_centers = []
    k_values = []
    for k in range(2, maximum_K):
        kmeans_model = KMeans(n_clusters = k)
        kmeans_model.fit(df)

        clusters_centers.append(kmeans_model.inertia_)
        k_values.append(k)
    return clusters_centers, k_values
```

```python
clusters_centers,k_values = find_best_clusters(scaled_data, 50)
```

```python
def generate_elbow_plot(clusters_centers, k_values):
    
    figure = plt.subplots(figsize = (12, 6))
    plt.plot(k_values, clusters_centers, 'o-', color = 'blue')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Cluster Inertia")
    plt.title("Elbow Plot of KMeans")
    plt.legend("blue = graph")
    plt.show()
```

```python
generate_elbow_plot(clusters_centers, k_values)
```

```python
kmeans_model = KMeans(n_clusters = 5)
```

```python
kmeans_model.fit(scaled_data)
```

```python
df["clusters"] = kmeans_model.labels_
```
